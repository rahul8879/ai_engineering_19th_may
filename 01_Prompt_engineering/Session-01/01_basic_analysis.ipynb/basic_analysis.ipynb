```python
def clean_income(x):
    x = x.replace('₹','')
    x = x.replace(',','')
    result = float(x)

    return result

data['Monthly Income (₹)'] = data['Monthly Income (₹)'].apply(clean_income)
```

```python
def predict(df,new_cusomter,k=3):
    distance= {}
    for index,row in df.iterrows():
        v1 = np.array([row['CIBIL Score'],row['Monthly Income (₹)']])
        v2 = np.array([new_cusomter[0],new_cusomter[1]])
        distance[index] = np.sqrt((v1[0]-v2[0])**2 + (v1[1]-v2[1])**2)

    # sort it 
    sorted_distance = sorted(distance.values(),reverse=False)
    return sorted_distance[:k]



```

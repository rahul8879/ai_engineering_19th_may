```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
load_dotenv()
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")
prompt =ChatPromptTemplate.from_messages([
    ('system', 'You are an expert support email classifier for a B2B SaaS company.'),
    ('human',  'Classify this email.\nSubject: {subject}\nBody: {body}\n\nReturn ONLY: Category | Urgency')
])

from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
```

```python
# Pydantic output 
chain = prompt | llm | parser 
chain.invoke({
    'subject': "Can't login after payment",
    'body': "Paid for annual plan last week, now can't login."
})
```

```python
# lets say you have 100 emails
# for loop ??

chain.batch([
    {
        'subject': "Can't login after payment",
        'body': "Paid for annual plan last week, now can't login."
    },
    {
        'subject': "Can't login after payment",
        'body': "Paid for annual plan last week, now can't login."
    },
    {
        'subject': "Can't login after payment",
        'body': "Paid for annual plan last week, now can't login."
    }
])
```

```python
from pydantic import BaseModel, Field
from typing import List, Optional, Literal
class EmailClassification(BaseModel):
    category: str = Field(description="One of: Billing, Technical, Feature Request, Spam, Other")
    urgency: str = Field(description="One of: High, Medium, Low")
    reasoning: str = Field(description="Step-by-step reasoning for this classification")

result =EmailClassification(category="Rahul", urgency="High", reasoning="...")
```

```python
class EmailClassificationresult(BaseModel):
    category: Literal["Billing", "Technical", "Feature Request", "Spam", "Other"] = Field(description="One of: Billing, Technical, Feature Request, Spam, Other")
    urgency: Literal["High", "Medium", "Low"] = Field(description="One of: High, Medium, Low")
    confidence : float = Field(
        ge=0.0, le=1.0,
        description="Confidence score for this classification")
    
    summary: str = Field(
        max_length=100,
        description="Short summary of the classification")
    esclate: Optional[bool] = Field(
        default=False,
        description="Should this email be escalated to the product team?"
    )
    reasoning: str = Field(description="Step-by-step reasoning for this classification")
```

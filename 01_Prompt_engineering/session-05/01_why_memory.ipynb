```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
load_dotenv()
```

```python
llm = ChatOpenAI(model="gpt-4o-mini")

call1 = llm.invoke("I received a billing complaint from John about invoice #1042.")
print(call1.content)
```

```python
call1 = llm.invoke("what was the invoice number.")
print(call1.content)
```

```python
llm.invoke("I received a billing complaint from John about invoice #1042. what was the invoice number.").content
```

```python
llm.invoke('what category does his comnplaints fall under')
```

```python
# manaul context stiching
user_turn_1 = "I received a billing complaint from John about invoice #1042. He was charged $500 instead of $250."
ai_turn_1 = llm.invoke(user_turn_1).content
print(ai_turn_1)

```

```python
user_turn_2 = "What category does his complaint fall under?"

stiched_prompt_turn_2 = f"""Previous conversation :
User: {user_turn_1}
AI: {ai_turn_1}
Now answer
User: {user_turn_2}"""

print(stiched_prompt_turn_2) #stiched_prompt_turn_2
```

```python
ai_turn_2 = llm.invoke(stiched_prompt_turn_2).content
print(ai_turn_2)
```
